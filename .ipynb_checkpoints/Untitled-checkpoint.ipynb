```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
shows1 = pd.read_csv("artifacts/shows1.csv")
shows2 = pd.read_csv("artifacts/shows2.csv")
shows = pd.concat([shows1, shows2], axis=0, ignore_index=True)

show_votes = pd.read_csv("data/show_votes.csv")
air_dates = pd.read_csv("data/air_dates.csv")
genres = pd.read_csv("data/genres.csv")
genre_types = pd.read_csv("data/genre_types.csv")
spoken_languages = pd.read_csv("data/spoken_languages.csv")
spoken_language_types = pd.read_csv("data/spoken_language_types.csv")
status = pd.read_csv("data/status.csv")
types = pd.read_csv("data/types.csv")
networks = pd.read_csv("data/networks.csv")
network_types = pd.read_csv("data/network_types.csv")
production_companies = pd.read_csv("data/production_companies.csv")
production_company_types = pd.read_csv("data/production_company_types.csv")
production_countries = pd.read_csv("data/production_countries.csv")
origin_country_types = pd.read_csv("data/origin_country_types.csv")
```

```python
# Most Popular TV Shows

popular_votes = show_votes[show_votes['vote_count'] > 10000]
shows.merge(popular_votes, on='show_id').sort_values('popularity', ascending=False).head(12)
```

```python
# Genre Popularity

temp = shows.merge(genres, on='show_id')
temp = temp.groupby('genre_type_id')['popularity'].mean().reset_index()
temp = temp.merge(genre_types, on='genre_type_id')
proportions = temp['popularity'].values
categories = temp['genre_name'].values
explode = [0 for i in range(len(categories))]
explode[16] = 0.1
explode[17] = 0.2

plt.figure(figsize=(8, 8))
plt.pie(proportions, labels=categories, autopct='%1.1f%%',explode=explode, startangle=140)
# Add a title
# plt.title('Distribution of Categories')

# Display the plot
plt.show()
```

```python
popular_languages = spoken_languages['spoken_language_type_id'].value_counts().head(12).reset_index()
categories = popular_languages.merge(spoken_language_types, on='spoken_language_type_id')['spoken_language_name'].values
values = popular_languages['count'].values

plt.figure(figsize=(10, 6))  # Adjust figure size if needed
sns.barplot(x=values, y=categories, palette='viridis')

plt.xlabel('Values')
plt.ylabel('Categories')
plt.title('Horizontal Bar Chart')
plt.show()
```

```python
temp = shows.merge(status, on='status_id')
temp = temp.groupby("status_name")['show_id'].count()

categories = temp.index
values = temp.values

sns.barplot(x=categories, y=values, palette='viridis')

plt.xlabel('Values')
plt.ylabel('Categories')
plt.title('Horizontal Bar Chart')
plt.xticks(rotation=90)

plt.show()
```

```python
temp = shows.merge(types, on='type_id')
temp = temp.groupby("type_name")['show_id'].count()

categories = temp.index
values = temp.values

sns.barplot(x=categories, y=values, palette='viridis')

# Add labels and title
plt.xlabel('Values')
plt.ylabel('Categories')
plt.title('Horizontal Bar Chart')
plt.xticks(rotation=90)

# Show the plot
plt.show()
```

```python
popular_networks = networks['network_type_id'].value_counts().head(7).reset_index()
categories = popular_networks.merge(network_types, on='network_type_id')['network_name'].values
values = popular_networks['count'].values


plt.pie(values, labels=categories, autopct='%1.1f%%', startangle=90)

# Draw a white circle at the center of pie to make it look like a donut
centre_circle = plt.Circle((0,0),0.70,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

plt.title('Donut Chart')
plt.axis('equal')
plt.show()
```

```python
# Production Company Insights
temp = production_companies['production_company_type_id'].value_counts().head(10).reset_index()
temp = temp.merge(production_company_types, on='production_company_type_id')

categories = temp['production_company_name'].values
values = temp['count'].values
```

```python
air_dates['date'] = pd.to_datetime(air_dates['date'])
```

```python
air_dates.info()
```

```python

temp = air_dates.merge(shows, on='show_id').groupby(['date', 'is_first'])['show_id'].count()
temp = temp.unstack()

fig, axs = plt.subplots(2, 1,sharex=False, figsize=(10, 8))

axs[0].plot(temp.index, temp[0])
axs[0].set_title('Line Plot 1')
axs[0].set_xlabel('Date')
axs[0].set_ylabel('Values')

axs[1].plot(temp.index, temp[1])
axs[1].set_title('Line Plot 2')
axs[1].set_xlabel('Date')
axs[1].set_ylabel('Values')

# Adjust layout
plt.tight_layout()

# Show plot
plt.show()
```

```python
temp = air_dates.merge(shows, on='show_id').merge(genres, on='show_id').merge(genre_types, on='genre_type_id')
temp = temp.groupby(['date', 'genre_name', 'is_first'])['show_id'].count()
temp = temp.unstack()
temp.reset_index()
```

```python
fig, axs = plt.subplots(2, 1,sharex=False, figsize=(10, 8))

for column in temp[0].columns:
    axs[0].plot(temp.index, temp[0][column])
axs[0].set_title('Line Plot 1')
axs[0].set_xlabel('Date')
axs[0].set_ylabel('Values')

for column in temp[1].columns:
    axs[1].plot(temp.index, temp[1][column])
axs[1].set_title('Line Plot 2')
axs[1].set_xlabel('Date')
axs[1].set_ylabel('Values')

# Adjust layout
plt.tight_layout()

# Show plot
plt.show()
```

```python
# production_countries = pd.read_csv("data/production_countries.csv")
# origin_country_types = pd.read_csv("data/origin_country_types.csv")

top_origin_countries_index = production_countries['origin_country_type_id'].value_counts().head(5).index
top_origin_countries = production_countries[production_countries['origin_country_type_id'].isin(top_origin_countries_index)]

temp = top_origin_countries.merge(origin_country_types, on='origin_country_type_id')
temp = temp.merge(shows, on='show_id')
temp = air_dates.merge(temp, on='show_id')
temp = temp.groupby(['date', 'origin_country_name'])['show_id'].count().unstack()
temp
```

```python
plt.figure(figsize=(12, 6))
for column in temp.columns:
    plt.plot(temp.index, temp[column], label=column)
plt.title('Line Plot with Dates on X-axis')
plt.xlabel('Date')
plt.ylabel('Values')
plt.grid(True)
plt.xticks(rotation=45)  # Rotate x-axis labels for better visibility
plt.tight_layout()
plt.legend()
plt.show()
```
